```python
class Perceptron: #codigo para operacion NOT
#le damos a la neurona su personalidad:
    def __init__(self, pesos, sesgo, tp):
        self.pesos = pesos #que tan importante consideras el valor del dato
        self.sesgo = sesgo ##donde se tiene que llegar para la activacion
        self.tp = tp ##(taza de aprendizaje) que tan rapido quieres que aprenda
    #la neurona hace las operaciones para saber si el valor es mayor o menor a cero
#not solo tiene una entrada a diferencia de las otras operaciones
    def funcion_activacion(self, z):
        if z >= 0:
            return 1
        else:
            return 0
    #trabaja la neurona con un peso
    def entrenamiento(self, x, y, epocas):
        for epoca in range(epocas):
            for i, dato in enumerate(x):
                x1 = dato[0]
                w1 = self.pesos[0]
                b = self.sesgo
          #aqui solo hay una entrada
                z = x1 * w1 + b
                y_pred = self.funcion_activacion(z)
                y_real = y[i]

                self.pesos[0] = w1 + self.tp * (y_real - y_pred) * x1
                self.sesgo = b + self.tp * (y_real - y_pred)

                print(f"dato {dato}, prediccion {y_pred}, pesos {self.pesos}, sesgo {self.sesgo}")
            print(f"fin de la epoca {epoca}")

# Datos para NOT
x = [[0], [1]]
y = [1, 0]
pesos = [-0.1]
sesgo = 0.1
tp = 0.1

p = Perceptron(pesos, sesgo, tp)
p.entrenamiento(x, y, 10)
```
