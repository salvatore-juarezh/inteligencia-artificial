```python
#paqueteria que vamos usar para hacer operacion XOR
import numpy as np

# Datos XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
##red neuronalnal
from tensorflow.keras.models import Sequential ##modelo de red neuro
from tensorflow.keras.layers import Dense ##creacion de capas

modelo = Sequential([
    Dense(4, activation='relu', input_dim=2), #agrega una capa de 4 neuronas. RELU, para no linealidad
    Dense(1, activation='sigmoid')
])#se agrega otra capa con 1 neurona, .SIGMO para salidas entre 1 y cero
```

```python
##copilar modelo. adam para optimizar el modelo
from tensorflow.keras.optimizers import Adam

# Tasa de aprendizaje deseada. 0.1 es el poder con el con el que optimizara
learning_rate = 0.1
adam_optimizer = Adam(learning_rate=learning_rate)
```

```python
modelo.compile(
    optimizer=adam_optimizer,
    loss='binary_crossentropy', #especifica la perdida
    metrics=['accuracy'] #metrica
)
#configura modelo para entrenar
```

```python
#inicio del entrenamiento
history = modelo.fit(
    X, y,
    epochs=25, #numero de veces que recorre el conjunto a entrenar
    verbose=1, #info que muestra durante entrenamiento
)
```

```python
##graficando la funcion perdida #tiene que ir hacia cero
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.title('Evolución de la pérdida (Loss)')
plt.xlabel('Época')
plt.ylabel('Binary Cross-Entropy')
plt.show()
```

```python
##evaluar modelo
loss, accuracy = modelo.evaluate(X, y)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

```python
##prediccion
pred = (modelo.predict(X) > 0.5).astype(int)
for i, (inp, p) in enumerate(zip(X, pred)):
    print(f"{inp} -> {p[0]}")
```
